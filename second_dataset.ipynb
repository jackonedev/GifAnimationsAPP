# INFORMACION SOBRE LA DESCARGA DE GIF's

Al guardar una animaciÃ³n como un archivo GIF, hay varios parÃ¡metros que puedes ajustar para controlar la calidad y el tamaÃ±o del archivo generado.

El parÃ¡metro mÃ¡s importante es **fps**, que define la velocidad de la animaciÃ³n en cuadros por segundo. Puedes ajustar este parÃ¡metro para hacer que la animaciÃ³n sea mÃ¡s rÃ¡pida o mÃ¡s lenta.

Otro parÃ¡metro importante es **dpi**, que controla la resoluciÃ³n de la animaciÃ³n. Aumentar este valor aumentarÃ¡ la resoluciÃ³n de la animaciÃ³n, lo que resultarÃ¡ en un archivo mÃ¡s grande y una descarga mÃ¡s lenta.

AdemÃ¡s, puedes ajustar el tamaÃ±o del archivo generado utilizando los parÃ¡metros **writer** y **bitrate**. writer controla el mÃ©todo utilizado para generar el archivo GIF, mientras que bitrate controla la tasa de bits utilizada para comprimir el archivo. Si utilizas un valor bajo para bitrate, el archivo resultante serÃ¡ mÃ¡s pequeÃ±o pero la calidad de la imagen serÃ¡ menor.

Los valores por defecto son:

fps: 10
dpi: 100
bitrate: None

El valor por defecto de fps es 10, lo que significa que la animaciÃ³n se reproducirÃ¡ a 10 cuadros por segundo. El valor por defecto de dpi es 100, lo que define la resoluciÃ³n de la imagen en puntos por pulgada. El valor por defecto de bitrate es None, lo que significa que se utilizarÃ¡ la configuraciÃ³n por defecto del writer para la compresiÃ³n del archivo.

    - La descarga de ani_1_3_2.gif tardÃ³ 16 minutos y 58.6 segundos
