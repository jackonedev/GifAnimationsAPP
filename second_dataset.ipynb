# INFORMACION SOBRE LA DESCARGA DE GIF's

Al guardar una animación como un archivo GIF, hay varios parámetros que puedes ajustar para controlar la calidad y el tamaño del archivo generado.

El parámetro más importante es **fps**, que define la velocidad de la animación en cuadros por segundo. Puedes ajustar este parámetro para hacer que la animación sea más rápida o más lenta.

Otro parámetro importante es **dpi**, que controla la resolución de la animación. Aumentar este valor aumentará la resolución de la animación, lo que resultará en un archivo más grande y una descarga más lenta.

Además, puedes ajustar el tamaño del archivo generado utilizando los parámetros **writer** y **bitrate**. writer controla el método utilizado para generar el archivo GIF, mientras que bitrate controla la tasa de bits utilizada para comprimir el archivo. Si utilizas un valor bajo para bitrate, el archivo resultante será más pequeño pero la calidad de la imagen será menor.

Los valores por defecto son:

fps: 10
dpi: 100
bitrate: None

El valor por defecto de fps es 10, lo que significa que la animación se reproducirá a 10 cuadros por segundo. El valor por defecto de dpi es 100, lo que define la resolución de la imagen en puntos por pulgada. El valor por defecto de bitrate es None, lo que significa que se utilizará la configuración por defecto del writer para la compresión del archivo.
